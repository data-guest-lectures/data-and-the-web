##Requests package for executing web requests [('HTTP for Humans')](http://docs.python-requests.org/en/latest/)

```python
import requests
```

##Now we can use the requests package to load the result into Python

```python
r = requests.get(url)
```

##This is formatted as a [dictionary](http://learnpythonthehardway.org/book/ex39.html)
